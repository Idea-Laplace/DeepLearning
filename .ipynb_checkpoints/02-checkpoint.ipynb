# 2
### Perceptron

```python
w1 = float(input('Set the value of weight1: '))
w2 = float(input('Set the value of weight2: '))
theta = float(input('Set the value of resistance: '))

def perceptron(x1, x2):
    if w1*x1 + w2*x2 > theta:
        return 1
    else:
        return 0

list = [(0,0), (1,0), (0,1), (1,1)]

for value in list:
    print(perceptron(*value))
    
```

# 2.1, 2.2
### AND, OR, NAND

```python
# 퍼셉트론을 응용하여 기본적인 게이트를 만들 수 있다.

def AND(x1, x2):
    if 0.5*x1 + 0.5*x2 > 0.9:
        return 1
    else:
        return 0

def OR(x1, x2):
    if 0.5*x1 + 0.5*x2 > 0.1:
        return 1
    else:
        return 0
    
def NAND(x1, x2):
    if -0.5*x1 - 0.5*x2 > - 0.9:
        return 1
    else:
        return 0

list_input = [(0,0), (1,0), (0,1), (1,1)]
list_gate = [AND, OR, NAND]
for func in list_gate:
    print(f'{str(func).split()[1]} gate')
    for input in list_input:
        print(f'Input:{input}, Output:', func(*input))
    print()
```

# 2.5
### XOR, Multi-layer perceptron

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    return AND(s1, s2)

print(f'{str(XOR).split()[1]} gate')
for input in list_input:
    print(f'Input:{input}, Intermediate:({NAND(*input)}, {OR(*input)}), Output:', XOR(*input))
```
