### 브로드 캐스트


```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
z = 5

#Element-wise
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print()
#Scalar operations
print(x/z)
print(x*z)
print(x+z)
print()


# Mulydimension array
A = np.array([[1, 2], [3, 4]])
B = np.array([[1, 3], [5, 7], [9, 11]])

print(A)
print(A.shape)
print(A.dtype)
print()

print(B)
print(B.shape)
print(B.dtype)
print()
#Broadcast
print(B*np.array([1, 2]))
print(B*np.array([[1, 2]]))
# 브로드캐스트 되는 방식은 그때 그때 확인할 것
# print(A*B) Not Broadcastable case
```

# 1.5.6
### 원소접근

```python
B = np.array([[1, 3], [5, 7], [9, 11]])
print(B[1])
print(B[0][1]) # [row][column]
print(B.flatten())
print(B) # flatten은 비파괴 함수

print(B.flatten() > 3) # 원소가 bool값으로 반환
print(B.flatten()[B.flatten() > 3]) # 조건에 맞는 인덱스를 찾아준다.
```

# 1.6.1
### matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1) #min, max, interval 조건에 맞춰 넘파이 어레이 객체를 생성
y = np.sin(x) # x interval 기준으로 y interval 도 결정됨
print(x)
print(type(x))
print(y)
print(type(y))
plt.plot(x,y) # x interval 기준으로 y도 interval 도 결정됨
plt.show()
```

# 1.6.2
## matplotlib.pyplot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1) #min, max, interval 조건에 맞춰 넘파이 어레이 객체를 생성
y1 = np.sin(x) # x interval 기준으로 y1 interval 도 결정됨
y2 = np.cos(x) # x interval 기준으로 y2 interval 도 결정됨

plt.plot(x, y1, 'o', label= 'sin') #plt.legend로 범례를 활성화했을 때 label 설정값이 나타남
plt.plot(x, y2, '+', label = 'cos')
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
plt.title('sin&cos') # 전체 이름
plt.legend() # 범례 표기
plt.show()
#help(plt)
```
